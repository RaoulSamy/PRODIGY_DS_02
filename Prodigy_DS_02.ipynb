# **TASK 2:- EXPLORATORY DATA ANALYSIS (EDA)**

The task here is to perform data cleaning and exploratory data analysis (EDA) on a dataset, explore the relationships between variables and identify patterns and trends in the data.

The dataset used here is Titanic Dataset from Kaggle.
https://www.kaggle.com/c/titanic/data

The 'train.csv' file is used from the dataset for this particular task.

**IMPORTING THE LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**IMPORTING THE DATASET**

```python
data = pd.read_csv('Titanic Dataset.csv')
data
```

**UNDERSTANDING THE DATASET AND DATA CLEANING**

```python
#To display the first 10 rows of the dataset
data.head(10)
```

```python
#To get basic details about the columns, its data types and number of non null values
data.info()
```

```python
#To understand the distribution or spread of the data
data.describe()
```

```python
#To find the number of missing values in each column
print(data.isnull().sum())
```

```python
#Distribution of Age

print("Minimum Age among passengers: ", data['Age'].min())
print("Maximum Age among passengers: ", data['Age'].max())
print("Mean Age of passengers: ", data['Age'].mean())
print("Number of missing values in Age column: ", data['Age'].isnull().sum())
```

**HANDLING MISSING VALUES**

Since *Age* is a very important factor for analysis and there are 177 missing values, let us fill it by using either mean or median. Since both the values are almost similar, let us take mean.

Since *Cabin* and *Embarked* Columns aren't of much significance, we discard them.

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
data.drop(['Cabin','Embarked'],1,inplace = True)
```

```python
print(data.isnull().sum())
```

**EDA**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=15, kde=True, color='orange')
plt.title('Distribution of Age among Titanic Passengers')
plt.show()
```

***Inference:-***

Majority of the passengers travelled are between 20 and 35 years.

```python
#Comparison between Gender and Survival
data[['Survived','Sex']].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Count by Sex')
plt.legend(['Not Survived','Survived'])
plt.show()
```

***Inference:-***

We can conclude that there were more survivors among female compared to male eventhough male population was high on the ship. More prefernce was given to females during the evacuation process.

```python
#Comparison between Passenger Class and Survival

plt.figure(figsize=(8, 6))

data['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', legend=True)
# plt.xlabel('Passenger Class')
# plt.ylabel('Number of Passengers')
plt.title('Distribution of Passengers in different Passenger Classes')
plt.show()

print()

sns.countplot(x='Pclass', hue='Survived', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Count by Passenger Class')
plt.legend(['Not Survived','Survived'])
plt.show()
```

***Inference:-***

Around half of the total passenger strength was in Class 3 and the survival rate in that class is less compared to the other classes. Among the 3 classes, Class 1 passengers survived the most.  

```python
#Comparison between Age and Passenger class
sns.boxplot(x='Pclass', y='Age', data=data)
plt.show
```

***Inference:-***

We can conclude that Class 1 had people of almost same age group i.e between 30 and 50. The other 2 classes had little bit diverse age groups, especially Class 3.

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(x='Age', hue='Survived', data=data, fill=True, common_norm=False)
plt.title('Age Distribution by Survival Status')
plt.legend(['Not Survived','Survived'])
plt.show()
```

***Inference:-***

We can conclude that majority of survivors as well as non-survivors are in the age range of 20 to 35.

```python
#Correlation Matrix
data[['Survived','Pclass','Age','Fare']].corr()
```

```python
#Visualize this correlation
plt.figure(figsize=(8, 6))
sns.heatmap(data[['Survived','Pclass','Age','Fare']].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

***Inference:-***

-> There is very weak negative correlation between *Age* and *Survived*, meaning higher the age, lesser the chance of survival.

-> There is negative correlation between *Fare* and *Class* and positive correlation between *Fare* and *Survived*, meaning as Class 1 has the highest fare and the fare decreases with the next class. Also means that, more money spent on ticket, more the survival chance.
